```python
import matplotlib.pyplot as plt
y=azerbaycan_veri['yıllar']
 
x=azerbaycan_veri['turist_sayısı']
plt.barh(y, x)
 
plt.ylabel("Yıllar")
 
plt.xlabel("Gelen Turist Sayıları (M)")
plt.title("Horizontal bar graph")
plt.show()
```

```python
import matplotlib.pyplot as plt
y=azerbaycan_veri['yıllar']
 
x=azerbaycan_veri['kisi_basina_yillik_gelir']
plt.barh(y, x)
 
plt.ylabel("Years")
 
plt.xlabel("GDP Per Capita")
plt.title("Horizontal bar graph")
plt.show()
```

```python
def value(x):
    rvalue  = np.round(x/100*azerbaycan_veri['turist_sayısı'].sum(), 0)
    return int(rvalue)
plot = azerbaycan_veri['turist_sayısı'].plot.pie(figsize=(7, 7),autopct=value)
plot.set_title('Turist Sayıları 1995-2020 yilları', fontsize=15)
pyplot.show()
```

```python
def value(x):
    rvalue  = np.round(x/100.*azerbaycan_veri['kisi_basina_yillik_gelir'].sum(), 0)
    return int(rvalue)
plot = azerbaycan_veri['kisi_basina_yillik_gelir'].plot.pie(figsize=(7, 7),autopct=value)
plot.set_title('Kişi Başına Düşen Yıllık Gelir 2020-1995 yılları', fontsize=15)
pyplot.show()
```
